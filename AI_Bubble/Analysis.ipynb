**Importing necessary libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Yahoo Finance API
import yfinance as yf
```

### Internet bubble

```python
internet_tickers = ['MSFT', 'INTC','ORCL', 'CSCO']
plt.figure(figsize=(15, 5))
for ticker in internet_tickers:
    data = yf.download(ticker, start="1999-01-01", end="2001-07-01")
    plt.plot(data['Adj Close'], label=ticker)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Adjusted close price')
plt.title('Adjusted close price of Internet stocks')
plt.show()
```

### META

```python
#get historical prices for META
meta = yf.Ticker("META")
meta_hist = meta.history(period="max")
#plot the historical prices
plt.figure(figsize=(12, 6))
plt.plot(meta_hist['Close'])
plt.title('META Historical Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### NVIDIA

```python
#get historical prices for NVIDIA
nvidia = yf.Ticker("NVDA")
#period between 2010-01-01 and 2021-01-01
nvidia_hist = nvidia.history(period="max")
#plot the historical prices
plt.figure(figsize=(12, 6))
plt.plot(nvidia_hist['Close'])
plt.title('NVIDIA Historical Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
